```python
# Label Encoding the Spectral Class Column 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Spectral Class"]=le.fit_transform(df["Spectral Class"])
df.head()
```

```python
df["Spectral Class"].unique()
```

```python
# Retrieving the features of the dataset
X=df.iloc[:,:-1].values
Y=df.iloc[:,-1].values
X
```

```python
# One Hot Encoding the Star color Column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[5])],remainder="passthrough")
X=np.array(ct.fit_transform(X))
print(X)
```

```python
# Standardizing the data
from sklearn import preprocessing
X=preprocessing.StandardScaler().fit(X).transform(X.astype(float))
```

```python
# Y=df.iloc[:,-1].values
Y
```

```python
# Splitting the dataset into training and test datasets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=1)    #25% : test data ; 75% : training data
print(X_train)
```

```python
print(X_test[:10])   # X_test
```

```python
print(Y_train)    # Y_train
```

```python
print(Y_test)   #Y_test
```

```python
# LOGISTIC REGRESSION CLASSIFIER
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression()
regressor.fit(X_train,Y_train)
y_pred=regressor.predict(X_test)
from sklearn import metrics
print(round(metrics.accuracy_score(Y_test,y_pred),2)*100,"%")
```

```python
# Classification Report for Logistic Regression Model
print(metrics.classification_report(Y_test,y_pred))
```

```python
# DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=0,max_depth=3)
clf_entropy.fit(X_train,Y_train)
y_pred2=clf_entropy.predict(X_test)
print(round(metrics.accuracy_score(Y_test,y_pred2),2)*100,"%")
```

```python
# Classification Report for Decision Tree Model
print(metrics.classification_report(Y_test,y_pred2))
```

```python
from sklearn import tree
fig=plt.figure(figsize=(25,20))
a=tree.plot_tree(clf_entropy,feature_names=X,class_names=["M","B","O","A","F","K","G"],filled=True)
a
plt.show()
```

```python
# RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100)
classifier.fit(X_train,Y_train)
y_pred3=classifier.predict(X_test)
print(round(metrics.accuracy_score(Y_test,y_pred3),2)*100,"%")
```

```python
# Classification Report for Random Forest Model
print(metrics.classification_report(Y_test,y_pred3))
```

```python
# K-NEAREST NEIGHBORS CLASSIFIER
from sklearn.neighbors import KNeighborsClassifier
k=15
knn_clf=KNeighborsClassifier(n_neighbors=k,metric="minkowski",p=2).fit(X_train,Y_train)
y_pred4=knn_clf.predict(X_test)
print(round(metrics.accuracy_score(Y_test,y_pred4),2)*100,"%")
```

```python
# Classification Report for Knn Model
print(metrics.classification_report(Y_test,y_pred4))
```

```python
# NAIVE BAYES CLASSIFIER (GAUSSIAN NB)
from sklearn.naive_bayes import GaussianNB
nb_clf=GaussianNB()
nb_clf.fit(X_train,Y_train)
y_pred5=nb_clf.predict(X_test)
print(round(metrics.accuracy_score(Y_test,y_pred5),2)*100,"%")
```

```python
# Classification Report for Naive Bayes(GaussianNB) Model
print(metrics.classification_report(Y_test,y_pred5))
```

```python
# Among the above Classification model Random Forest Classification model has the highest accuracy.Therefore for this Stars Classification dataset,
# Random Forest Classifier is apt.
```
